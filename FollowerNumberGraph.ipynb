```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

```python
path = "/Users/rachel/KAIST/CT/SocialNetworkAnalysis/gitRep/FirstTrial/0516/"
filename_list = ["macron_0510_500", "lepen_0510_500", "melenchon_0510_500"]

#log scale
def follower_log_list(path, filename):
    follower_num_list = []
    following_num_list = []
    tweets_filename = path + filename + "tweets.csv"
    tweets = open(tweets_filename,'r')
    rdr = csv.reader(tweets)
    next(rdr)
    for line in rdr:
        followers = int(line[-5])
        followings = int(line[-6])
        if followers==0:
            follower_num_list.append(0)
        else: 
            follower_num_list.append(math.log10(followers))
        if followings==0:
            following_num_list.append(0)
        else:
            following_num_list.append(math.log10(followings))
    return following_num_list, follower_num_list

#plain scale
def follower_num_list(path, filename):
    follower_num_list = []
    following_num_list = []
    tweets_filename = path + filename + "tweets.csv"
    tweets = open(tweets_filename,'r')
    rdr = csv.reader(tweets)
    next(rdr)
    for line in rdr:
        followers = int(line[-5])
        followings = int(line[-6])
        if followers==0:
            follower_num_list.append(0)
        else: 
            follower_num_list.append(followers)
        if followings==0:
            following_num_list.append(0)
        else:
            following_num_list.append(followings)
    return following_num_list, follower_num_list

def scatter_follower_nums(path, filename_list):
    for filename in filename_list:
        followings, followers = follower_num_list(path, filename)
        plt.scatter(followings, followers)
        a, b = np.polyfit(followings, followers, 1) 
        x_minmax = np.array([min(followings), max(followings)])
        fit_y = x_minmax * a + b
        plt.plot(x_minmax, fit_y, color='orange')
        print(f"y = {round(a,3)}x + ({round(b,3)})")
        
def scatter_follower_logs(path, filename_list):
    for filename in filename_list:
        followings, followers = follower_log_list(path, filename)
        plt.scatter(followings, followers)
        a, b = np.polyfit(followings, followers, 1)
        x_minmax = np.array([min(followings), max(followings)])
        fit_y = x_minmax * a + b
        plt.plot(x_minmax, fit_y, color='orange')
        print(f"y = {round(a,3)}x + ({round(b,3)})")
```
